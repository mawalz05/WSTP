```python
from sklearn import preprocessing

# Normalizing the quantitative variables so they are on the sam scale
env_scaled_ = pd.DataFrame(preprocessing.scale(env_quant), columns = ['rec_per', 'org_per', 'waste_per', 'energy_ren_per'])
econ_scaled_ = pd.DataFrame(preprocessing.scale(econ_quant), columns = ['rec_per', 'energy_ren_per', 'research_per'])
social_scaled_ = pd.DataFrame(preprocessing.scale(social_quant), columns = ['employees', 'phil_emp_ratio'])

# Adding back the scaled quantitative variables to the categorical variables
env_scaled = pd.concat([env_scaled_, env_qual], axis = 1)
econ_scaled = pd.concat([econ_scaled_, econ_qual], axis = 1)
social_scaled = pd.concat([social_scaled_, social_qual], axis = 1)

# Examining the social scaled scores
social_scaled
```

```python
# Import libraries
import prince
import pprint

#https://github.com/MaxHalford/prince


## Instantiate FAMD (Factor Analysis for Mixed Data from prince module) object
famd = prince.FAMD(
    n_components=1, # How many dimensions we want to keep
    n_iter=10, # Number of iterations of SVD
    copy=True, # Always set to True according to the Prince documentation
    check_input=True, # Always set to True according to the Prince documentation
    engine='auto',       ## Can be "auto", 'sklearn', 'fbpca'
    random_state=42) # Make it reproducible with a random seed

## Fit FAMD object to social data 
famd = famd.fit(social_scaled) 

# Extracting the scores from the algorithm
social_scores = famd.row_coordinates(social_scaled)

social_scores = (social_scores-social_scores.min())/(social_scores.max()-social_scores.min()) #Normalizing the values on a 0 to 1 scale
social_scores = social_scores*100 # Readjusting the scaled values to 0 to 100
social_scores = (social_scores/2) + 50 # Making the baseline score 50

print("Amount of variance explained by first component: ", famd.explained_inertia_) # Amount of variance explained with this component
print("Eigenvalue for first component: ", famd.eigenvalues_) # The Eigenvalues obtained through SVD
print("Total inertia first component: ", famd.total_inertia_) # Can be used in conjuction with the eigenvalues to get explained variance

# Verifying for accuracy
famd.eigenvalues_/famd.total_inertia_ == famd.explained_inertia_
```

```python
## Instantiate FAMD object for environment variables
famd = prince.FAMD(
    n_components=1,
    n_iter=10,
    copy=True,
    check_input=True,
    engine='auto',       ## Can be "auto", 'sklearn', 'fbpca'
    random_state=42)

## Fit FAMD object to data 
famd = famd.fit(env_scaled) 

# Extracting the scores from the algorithm
env_scores = famd.row_coordinates(env_scaled)*-1 # Multiplying by negative one because the direction is flipped.

env_scores = (env_scores-env_scores.min())/(env_scores.max()-env_scores.min()) #Normalizing the values on a 0 to 1 scale
env_scores = env_scores*100 # Readjusting the scaled values to 0 to 100
env_scores = (env_scores/2) + 50

print("Amount of variance explained by first component: ", famd.explained_inertia_) # Amount of variance explained with this component
print("Eigenvalue for first component: ", famd.eigenvalues_) # The Eigenvalues obtained through SVD
print("Total inertia first component: ", famd.total_inertia_) # Can be used in conjuction with the eigenvalues to get explained variance

# Verifying for accuracy
famd.eigenvalues_/famd.total_inertia_ == famd.explained_inertia_
```

```python
## Instantiate FAMD object for economic variables
famd = prince.FAMD(
    n_components=1,
    n_iter=10,
    copy=True,
    check_input=True,
    engine='auto',       ## Can be "auto", 'sklearn', 'fbpca'
    random_state=42)

## Fit FAMD object to data 
famd = famd.fit(econ_scaled) 

# Extracting the scores from the algorithm
econ_scores = famd.row_coordinates(econ_scaled)

econ_scores = (econ_scores-econ_scores.min())/(econ_scores.max()-econ_scores.min()) #Normalizing the values on a 0 to 1 scale
econ_scores = econ_scores*100 # Readjusting the scaled values to 0 to 100
econ_scores = (econ_scores/2) + 50

print("Amount of variance explained by first component: ", famd.explained_inertia_) # Amount of variance explained with this component
print("Eigenvalue for first component: ", famd.eigenvalues_) # The Eigenvalues obtained through SVD
print("Total inertia first component: ", famd.total_inertia_) # Can be used in conjuction with the eigenvalues to get explained variance

# Verifying for accuracy
famd.eigenvalues_/famd.total_inertia_ == famd.explained_inertia_
```

```python
# Adding the scores to the final dataframe
final['env_scores'] = env_scores
final['econ_scores'] = econ_scores
final['social_scores'] = social_scores

# Creating the sustainability score as the average from the three pca parameters
final['sustainability_score_avg'] = (final['env_scores'] + final['econ_scores'] + final['social_scores'])/3

# Creating the sustainability score using WSTP weights
final['sustainability_631weight'] = (final['env_scores']*6 + final['econ_scores']*3 + final['social_scores'])/10

# Selecting the variables we want for the final dataset.
final = final[['company', 'survey_taker', 'product_name', 'env_scores', 'econ_scores', 'social_scores', 'sustainability_score_avg', 'sustainability_631weight']]
final
```
